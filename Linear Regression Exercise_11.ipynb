### Feature Engineering and Scaled Data 

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df['State'] = le.fit_transform(df['State'])
```

```python
df['State'].head(5)
```

```python
x=df.drop('Profit',axis=1)
x.head(1)
```

```python
y=df['Profit']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.20,random_state=45)
```

```python
x_train.head(2)
```

```python
y_train.tail(2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_test_pred=model.predict(x_test)
y_train_pred=model.predict(x_train)
y_test_pred[:5]
```

```python
y_test[:5]
```

```python
model.score(x_train,y_train)
```

### Analysis on model
  As you can see our predictive model has sored more than 90% and also by seeing the result of y_test_pred and
  y_test their is not much difference.
 *So this is the best fit for our model*

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_test_pred)
print('r2 score:',r2)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
```

```python
print(f"Training R: {r2:.3f}, RMSE: {rmse_train:.2f}")
print("Mean Squared Error:",mean_squared_error(y_test,y_test_pred))
```

# Model Performance

### R Score: 0.9836
The model explains **98.36% of the variability** in the target variable. This indicates a **best fit** and suggests that the predictors are capturing the majority of the relationship with the target variable.  

### Root Mean Square Error (RMSE): 9503.55
On average, the models predictions deviate from the actual values by approximately **9503 units**.  
- The magnitude of RMSE should be compared with the **range of the target variable**.  
- If the range is large (e.g., target values vary between tens or hundreds of thousands), this error may be **acceptable**.  
- If the range of the target variable is narrow, the error could indicate **room for improvement** in the model.

# Residual Plot Analysis

## Observations:

1. **Random Scatter**:
   - The residuals are scattered above and below the red dashed line (residual = 0), which is a good indication that the model doesn't have a systematic bias.  

2. **Presence of Patterns**:
   - There seems to be a slight trend where residuals increase or decrease with predicted values, especially at the lower and upper ends of the plot.  
   - This suggests **heteroscedasticity**, where the error variance changes across the range of predicted values.

3. **Outliers**:
   - A few residuals are significantly far from the zero line (e.g., above 5,000 or below -5,000).  
   - These outliers could indicate areas where the model struggles to make accurate predictions.

4. **Range of Residuals**:
   - Residuals range approximately between **-6,000 to 8,000**, indicating some significant errors in predictions.

```python
from sklearn.linear_model import Ridge, Lasso

ridge = Ridge(alpha=1.0)
ridge.fit(x_train, y_train)
r2_ridge = r2_score(y_test, ridge.predict(x_test))

# Lasso Regression (L1)
lasso = Lasso(alpha=0.01)
lasso.fit(x_train, y_train)
r2_lasso = r2_score(y_test, lasso.predict(x_test))

print(f"Ridge R: {r2_ridge:.3f}")
print(f"Lasso R: {r2_lasso:.3f}")
```

### Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
# Add a constant for VIF calculation
X_with_const = sm.add_constant(x)

# Calculate VIF
vif = pd.DataFrame()
vif['Features'] = x.columns
vif['VIF'] = [variance_inflation_factor(X_with_const.values, i) for i in range(1, X_with_const.shape[1])]

print(vif)
```
